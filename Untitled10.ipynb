```python
import astropy.units as u
from astropy.coordinates import SkyCoord, Angle
import numpy as np
import matplotlib.pyplot as plt
```

```python
from gammapy.image.models.core import SkyEllipse
from gammapy.maps import Map, WcsGeom
```

## Verify the application of the IAU convention regarding the position angle computation

```python
def plot_ellipse(ax, theta, frame, coordsys):
    model = SkyEllipse("2 deg", "2 deg", "1 deg", 0.9, theta, frame=frame)
    
    geom = WcsGeom.create(binsz=0.01, width=(3, 3), skydir=(2, 2), coordsys=coordsys, proj="AIT")
    coords = geom.get_coord()
    lon = coords.lon * u.deg
    lat = coords.lat * u.deg
    
    vals = model(lon, lat)
    
    skymap = Map.from_geom(geom, data=vals.value)
    _, ax, _ = skymap.smooth("0.05 deg").plot(ax=ax)
```

A SkyEllipse rotating counter-clockwise in the direction of positive right ascensions:

```python
geom = WcsGeom.create(binsz=0.01, width=(3, 3), skydir=(2, 2), coordsys='CEL', proj="AIT")
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,10), subplot_kw={'projection': geom.wcs})

plot_ellipse(ax=axes[0], theta='0 deg', frame='icrs', coordsys='CEL')
plot_ellipse(ax=axes[1], theta='20 deg', frame='icrs', coordsys='CEL')
plot_ellipse(ax=axes[2], theta='40 deg', frame='icrs', coordsys='CEL')

ref_points = SkyCoord(np.arange(-1, 1, 0.001), 0, unit='deg', frame='icrs')
```

The same, displayed in Galactic coordinates this time: 

```python
skydir_gal = SkyCoord(2*u.deg, 2*u.deg, frame='icrs').transform_to('galactic')

geom = WcsGeom.create(binsz=0.01, width=(3, 3), skydir=(skydir_gal.l.value, skydir_gal.b.value), coordsys='GAL', proj="AIT")
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,10), subplot_kw={'projection': geom.wcs})

plot_ellipse(ax=axes[0], theta='0 deg', frame='galactic', coordsys='GAL')
plot_ellipse(ax=axes[1], theta='20 deg', frame='galactic', coordsys='GAL')
plot_ellipse(ax=axes[2], theta='40 deg', frame='galactic', coordsys='GAL')
```

## Check the plot produced in the example docstring

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from gammapy.image.models.core import SkyEllipse
from gammapy.maps import Map, WcsGeom

# The reference for the position angle, indipendently of the
# chosen frame, is the direction of the North Galactic Pole
model = SkyEllipse("2 deg", "2 deg", "1 deg", 0.8, "20 deg", frame='icrs')

m_geom = WcsGeom.create(binsz=0.01, width=(3, 3), skydir=(2, 2), coordsys="CEL", proj="AIT")
coords = m_geom.get_coord()
lon = coords.lon * u.deg
lat = coords.lat * u.deg
vals = model(lon, lat)
skymap = Map.from_geom(m_geom, data=vals.value)

_, ax, _ = skymap.smooth("0.05 deg").plot()
plt.show()
```
