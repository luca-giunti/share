In this nb, I compare the PSF of a single HESSIU run (ID: 132122) between the FR and HD FITS prods

```python
import gammapy
gammapy.__version__
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import astropy.units as u
from astropy.coordinates import Angle
```

```python
from gammapy.analysis import Analysis, AnalysisConfig
```

```python
OBS_ID = 132122
```

# HAP-FR HE prod

```python
BASE_DIR = "/home/luca/HAP_fr_new_irfs/he_stereo_Prod17_Calib0834_thsq64/he_stereo_Prod17_Calib0834_thsq64"
```

```python
# Select one observation
config = AnalysisConfig()
config.observations.datastore = BASE_DIR
config.observations.obs_ids = [OBS_ID]
analysis = Analysis(config)
analysis.get_observations()
```

```python
obs = analysis.observations[0]
```

```python
obs.obs_id
```

```python
obs.psf.plot_containment_vs_energy(fractions=[0.68, 0.99], thetas=Angle([0, 1] * u.deg))
```

```python
print(obs.psf.info())
```

```python
obs.psf.to_energy_dependent_table_psf(theta=0 * u.deg).rad.to("deg").max()
```

```python
obs.psf.to_energy_dependent_table_psf(theta=0 * u.deg).plot_psf_vs_rad(energies=[2, 10, 100] * u.TeV)
plt.xlim(0.0, 1.5)
```
