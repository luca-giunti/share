## Check the consistency with `regions.EllipseSkyRegion`

```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from gammapy.image.models.core import SkyEllipse
from gammapy.maps import Map, WcsGeom

m_geom = WcsGeom.create(binsz=0.01, width=(3, 3), skydir=(2, 2), coordsys="GAL", proj="AIT")
model = SkyEllipse("2 deg", "2 deg", "1 deg", 0.8, "20 deg", frame='galactic')


coords = m_geom.get_coord()
lon = coords.lon * u.deg
lat = coords.lat * u.deg
vals = model(lon, lat)
skymap = Map.from_geom(m_geom, data=vals.value)

_, ax, _ = skymap.smooth("0.05 deg").plot()

# Overlay astropy ellipse
skydir = SkyCoord(2*u.deg, 2*u.deg, frame='galactic')
semi_major = 1 * u.deg
semi_minor = semi_major * np.sqrt(1 - 0.8**2)
width = 2*semi_minor
height = 2*semi_major
astropy_ellipse = EllipseSkyRegion(skydir, width, height, angle=20 * u.deg)
astropy_pix = astropy_ellipse.to_pixel(wcs=m_geom.wcs)
astropy_pix.plot(ax=ax, ls="--", edgecolor="green", lw=3)

plt.show()
```
